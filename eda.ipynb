# Packet Count

```python
def pckt_count(a):
    a = a.split(';')
    a = [int(i) for i in a[:-1]]
    return len(a)
```

```python
run_200_300['pckt_ct'] = run_200_300['packet_times'].apply(pckt_count)
```

```python
figure(figsize=(12, 6), dpi=80)
plt.plot(run_200_300['Time'], run_200_300['pckt_ct'])
plt.xlabel('Time (s)')
plt.ylabel('Bytes')
plt.title('Packets per second -- 1->2Bytes (200/300)')
```

```python
run_200_1000['pck_ct'] = run_200_1000['packet_times'].apply(pckt_count)

figure(figsize=(12, 6), dpi=80)
plt.plot(run_200_1000['Time'], run_200_1000['pck_ct'])
plt.xlabel('Time (s)')
plt.ylabel('Bytes')
plt.title('Packets per second -- 1->2Bytes (200/1000)')
```

```python
run_200_25000['pck_ct'] = run_200_25000['packet_times'].apply(pckt_count)

figure(figsize=(12, 6), dpi=80)
plt.plot(run_200_25000['Time'], run_200_25000['pck_ct'])
plt.xlabel('Time (s)')
plt.ylabel('Bytes')
plt.title('Packets per second -- 1->2Bytes (200/25000)')
```

```python
run_200_5000['pck_ct'] = run_200_5000['packet_times'].apply(pckt_count)

figure(figsize=(12, 6), dpi=80)
plt.plot(run_200_5000['Time'], run_200_5000['pck_ct'])
plt.xlabel('Time (s)')
plt.ylabel('Bytes')
plt.title('Packets per second -- 1->2Bytes (200/5000)')
```
